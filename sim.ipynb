```python
import numpy as np
import agent
import analyse
import visualise
```

```python
tmax = 100 #duration of the simulation
length = 128 # dimension of simulation (for 1D set to 1)
height = 20 # second dimension of the simulation
sigma = 0.2 # standard deviation of noise, or natural frequency distribution
eta = 0.2 # non linear coupling coefficient
bc = 'fix' # boundary conditions in x, can be 'fix' for fixed boundary values, 
            # 'grad' for defined gradient values, or 'periodic'
grad = [0,0] # values of the gradient boundary condition if this is chosen. [1,1], [-1,1], [1,-1],[0,0] are labelled A,B,C,D in filenames
dim = 'q2D' # dimension label for data organisation
```

```python
# for custom model with parameters set above
agent.setup()
model = agent.Kuramoto(tmax, height, length, sigma, eta, bc, grad,dim)
if length == 1:
    model.solve()
else:
    model.solve(dim=2)
model.save() # raw data file saved with date as title
```
